```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = 'colony.jpeg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```
